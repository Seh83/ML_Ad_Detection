```python
import numpy as np
import pandas as pd
```

```python
#data = pd.read_csv("./ad-dataset/ad.data", header=None, low_memory=False, sep=",")
data = pd.read_csv("./ad-dataset/ad.data")
```

```python
#ad.
#nonad.
data.head(20)
```

```python
data['ad.'].replace(("ad.", "nonad."),(1,0), inplace=True)
```

```python
data.head(5)
```

```python
def toNum(cell):
    try:
        return np.float(cell)
    except:
        return np.nan
```

```python
def seriesToNum(series):
    return series.apply(toNum)
```

```python
newerData = data.iloc[0:,0:].apply(seriesToNum)
```

```python
newerData.head()
```

```python
newerData.dropna(inplace=True)
```

```python
newerData.head()
```

```python
#Split into TrainData and TarinLabel

train_data = newerData.iloc[0:,0:-1]
train_labels = newerData['ad.']


```

```python
train_data.head()
```

```python
train_labels.head()
```

```python
#Import LinerSVC
from sklearn.svm import LinearSVC
```

```python
clf = LinearSVC()

clf.fit(train_data[100:2300], train_labels[100:2300])
```

```python
#Prediction If the ad. is an Ad or NonAd
clf.predict(train_data.iloc[12].reshape(1,-1))
```
